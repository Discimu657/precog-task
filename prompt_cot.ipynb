```python
import os
import groq
import requests
from dotenv import load_dotenv 
load_dotenv()
```

```python
from groq import Groq
client=Groq(
    api_key=os.environ.get("GROQ_API_KEY")   
)
chat_completion=client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
       {
           "role":"system",
            "content":"You are a helpful model"
           
       },
       {
         "role":"user",
         "content":"What are local llms how do they differ from general ones?"   
       }
    ]
)
def get_chat_completion(prompt):
     model="llama-3.3-70b-versatile"
     messages=[
       {
           "role":"system",
            "content":"You are a helpful model"
           
       },
       {
         "role":"user",
         "content":prompt   
       }
    ]
     chat_completion=client.chat.completions.create(
          model=model,
          messages=messages,
          temperature=0
     )
     return chat_completion.choices[0].message.content
     
```

```python
prompt="""
You are given the following problem:
{
  "problem_id": "001",
  "initial_string": "'R'=w!(~[",
  "transitions": [
    { "src": "'HV", "tgt": "" },
    { "src": "X%[", "tgt": "" },
    { "src": "DIb/", "tgt": "" },
    { "src": "@t~", "tgt": "DIb/%" },
    { "src": "iZ^-", "tgt": "HVX@t" },
    { "src": "Rs", "tgt": "iZ^'H" },
    { "src": "=w!(", "tgt": "HVsV-" }
  ]
}
Your task is to arrive at a final  blank string from applying the rules given. If the rule is say X->Y then you can replace the first instance  of  X with Y(total Y needs to be removed else rule can't be applied), so if Y is blank you can remove X, you need to fill the rules in an index format(0 for first transition) in the solution format I provide

Please provide a **chain-of-thought** explanation where you:
1. Examine the initial string and each transition.
2. Describe the reasoning behind each transformation step.
3. Detail how you decide the order of applying transitions.
4. Step through your linear reasoning until you derive the final solution.

Finally, output the final solution as an array.


"""
chat_completion=get_chat_completion(prompt)
print(chat_completion)
```
